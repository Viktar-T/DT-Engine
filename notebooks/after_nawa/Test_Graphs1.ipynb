```python
%matplotlib qt
import numpy as np
import pandas as pd
import itertools
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel('1600Nn obc ON _ 2018-12-06_ORIGEN.xlsx')
```

```python
# Harmonize_time --- concat()
df1 = df.iloc[:, [58, 59]].dropna()
df2 = df.iloc[:, [54, 55]].dropna()
df3 = df.iloc[:, [52, 53]].dropna()
df4 = df.iloc[:, [144, 145]].dropna()

df1 = df1.drop_duplicates([df1.columns[0]])
df2 = df2.drop_duplicates([df2.columns[0]])
df3 = df3.drop_duplicates([df3.columns[0]])
df4 = df4.drop_duplicates([df4.columns[0]])

df1 = df1.set_index(df1.columns[0])
df2 = df2.set_index(df2.columns[0])
df3 = df3.set_index(df3.columns[0])
df4 = df4.set_index(df4.columns[0])

#df1.index.name = 'czas'
#df2.index.name = 'czas'
frame = [df1, df2, df3, df4]

df_concat = pd.concat(frame, axis=1)
df_concat.fillna(method='ffill', inplace=True)
res_filtered = df_concat
df_concat
```

```python
res_filtered
```

```python
div_f = 0.05
lst_f = np.arange(0.05, 4, 0.01)
#lst_f = map(lambda x: x * 0.01, [x for x in range(5, 400, 1)])
min_time_f = 1500
indices = []
#f = 0.95
for f in lst_f:
    div_min = f - div_f
    div_max = f + div_f    
    index = res_filtered[(res_filtered['Zużycie paliwa średnie[g/s]'] >= div_min) & 
                     (res_filtered['Zużycie paliwa średnie[g/s]'] <= div_max)].index
    if len(index) > min_time_f:
        indices += list(index)


drop_ind = list(set(res_filtered.index) - set(indices))
res_filtered.drop(drop_ind, inplace=True)
res_filtered
```

```python
# filtering torque. Last value has different range. Use closure (замыкания функции) insted of two for loops
lst_torque = [1, 25, 50, 75, 100, 125, 150, 175, 200, 225, 250, 275, 300, 325] # without last point
lst_last = [i for i in range(330, 370, 5)]   # list_torque to find last point
div = 1  # diviation
min_time_experiment = 100

for tq in lst_torque:
    div_min = tq - div
    div_max = tq + div
    index = res_filtered[(res_filtered['Moment obrotowy[Nm]'] >= div_min) & 
                         (res_filtered['Moment obrotowy[Nm]'] <= div_max)].index
    if len(index) > min_time_experiment:
        res_filtered.loc[index, 'Torque'] = tq

for tq in lst_last:
    div_min = tq - div
    div_max = tq + div
    index = res_filtered[(res_filtered['Moment obrotowy[Nm]'] >= div_min) & 
                         (res_filtered['Moment obrotowy[Nm]'] <= div_max)].index
    if len(index) > min_time_experiment:
        res_filtered.loc[index, 'Torque'] = tq        
        index_last = index[-1]
    
res_filtered.drop(res_filtered[res_filtered.index > index_last].index, inplace=True) # drop tail of the the data
res_filtered.dropna(inplace=True)

grouped_res = res_filtered.groupby('Torque')  # group by 'Torque'
dict_gr_res = dict(list(grouped_res))         # get dict --- dictionary 1: df1, 25: df2, 50: df3, ...

res_filtered
```
