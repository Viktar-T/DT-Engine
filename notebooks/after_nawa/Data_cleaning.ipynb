```python
class Table_main:
    """ Class 
        1. def __init__(self, address):  
        5. def get_column_names:   --- return self.df_cname - names with index of all columns in df
        3. def harmonize_time(self): --- columns have different time steps - ['Czas [ms].26', 'Moc[kW]'] 
        4. def filter_data(self): --- Filtering Data
            4.1 def filter_rotation(self):
            4.2 def filtering_torque(self, lst):
            4.3 def filter_fuel(self): 
    """
    parameters_all = ['Ciś. pow. za turb.[Pa]', 'ECT - wyjście z sil.[°C]', 'MAF[kg/h]', 'Moc[kW]', 
                      'Moment obrotowy[Nm]', 'Obroty[obr/min]', 'Temp. oleju w misce[°C]', 
                      'Temp. pal. na wyjściu sil.[°C]', 'Temp. powietrza za turb.[°C]', 
      'Temp. spalin 1/6[°C]', 'Temp. spalin 2/6[°C]', 'Temp. spalin 3/6[°C]', 'Temp. spalin 4/6[°C]',  # count average
                      'Zużycie paliwa średnie[g/s]', 
      'Ciśnienie atmosferyczne[hPa]', 'Temp. otoczenia[°C]', 'Wilgotność względna[%]'] # lst_for_correction
    
    lst_for_correction = ['Ciśnienie atmosferyczne[hPa]', 'Temp. otoczenia[°C]', 'Wilgotność względna[%]']
        
    # for def filter_data(self) --- filtering rotation
    div_r = 10
    
    # for def filter_rotation_ex(self): --- used only for external speed characteristics
    #lst_rotation = [1300, 1400, 1500, 1600, 1700, 1800, 1900, 2000, 2100, 2200, 2300, 2400]
    lst_rotation = [2400, 2300, 2200, 2100, 2000, 1900, 1800, 1700, 1600, 1500, 1400, 1300]
    
    # for def filtering torque
    lst_torque = [1, 25, 50, 75, 100, 125, 150, 175, 200, 225, 250, 275, 300, 325] # without last point
    lst_last = [i for i in range(330, 370, 5)]   # list_torque to find last point
    div_t = 1  # diviation of the torque
    min_time_experiment = 100
    
    # for def filter_fuel(self):
    div_f = 0.04
    #lst_f = np.arange(0.05, 4, 0.01)
    #min_time_f = 700
        
    # def filter_first_last --- cut the begining and end of each torque level
    first_drop = 200
    last_drop = 50
    
        
    # charac_type='load' or 'ex_speed'; rotation=1600,1800,..., 2400-rated  
    # fuel_means=[] --- we are choosing means for each mode by hands 
    def __init__(self, address, address_eco, charac_type='load', rotation=2400, fuel_means=[], fuel_name='Diesel'):  
        '''        !!!!!!!!!!!!!!!!!!!!!!      '''
        
        self.df = pd.read_excel(address)    # !!!!!!!!!!!!!!!!!!
        self.df_eco = pd.read_excel(address_eco)
        #self.df = df
        #self.df = df_ex
        #self.df_eco = df_ex_eco
        self.rotation = rotation  # used ONLY for load characteristics --- def filter_rotation(self): 
        self.charac_type = charac_type
        self.fuel_means = fuel_means   # used in def filter_fuel(self):
        self.fuel_name = fuel_name # used in self.add_fuel_prop()

        # get lists of parameters. Check whether parameters_all are in the current df.
        self.get_lst_param()
        # def harmonize_time(self):
        self.harmonize_time()
        # def filter_data(self):        
        self.filter_data()
        # def transform(self):
        self.transform()
        #def add_ecology(self)
        self.add_ecology()
        # add properties of the fuels
        self.add_fuel_prop()
        # group by 'Torque' or 'Rotation' --- and --- get dict
        self.get_dict()
        
        self.write_to(f'res_class_Table_main\\tr_{fuel_name}.csv')

      
    def get_lst_param(self):
        self.parameters = []
        self.correction = []
        self.lst_columns = list(self.df.columns)
        for par in self.parameters_all:
            if par in self.df.columns:                
                self.index_df_par = self.lst_columns.index(par)
                self.parameters.append(list([self.lst_columns[self.index_df_par - 1], 
                                            self.lst_columns[self.index_df_par]])) 
        print('parameters:', *self.parameters, sep='\n', end='\n\n')
        
    
    # according to the time merge() rowes. I could not able to create a loop.
    def harmonize_time(self):
        self.frame = []
        for s in self.parameters:          
            # delet rows with fast sensors.  
            self.df_x = self.df[s].dropna()
            self.df_x = self.df_x.drop_duplicates([self.df_x.columns[0]])
            self.df_x = self.df_x.set_index(self.df_x.columns[0])            
            self.frame.append(self.df_x)
            
        self.res_concat = pd.concat(self.frame, axis=1)
        print('time was harmonized --- def harmonize_time(self):')
        print(f'{len(self.res_concat)} - whole amount or rowes in DataFrame', end='\n\n')
        #return self.res_concat

    ''' !!!!!!!!!! FILTERING !!!!!!!!!!! ''' 
    # Filtering the Data, # drop first rowes --- use if 
    def filter_data(self):
        self.dict_all_filtures = {'def filter_rotation(self):': False,     # dictionary of all filtures
                                  '    1.amount of rowes --- ': np.nan, 
                                  'def filter_torque(self, lst):': False,
                                  '    2.amount of rowes --- ': np.nan,
                                  'def filter_rotation_ex(self):': False,
                                  '    3.amount of rowes --- ': np.nan,
                                  'def filter_torque_ex(self):': False,
                                  '    4.amount of rowes --- ': np.nan,
                                  'def filter_fuel(self):': False,
                                  '    5.amount of rowes --- ': np.nan,
                                  'def filter_first_last(self):': False,
                                  '    6.amount of rowes --- ': np.nan}
        
        self.res_filtered = self.res_concat.fillna(method='ffill')
        
        
        if self.charac_type == 'load':
            # filtering rotation frequency
            self.filter_rotation()
            # filtering torque                
            self.filter_torque(self.lst_torque)   # filter without last point
            self.filter_torque(self.lst_last)   # filter to find last point
            
        elif self.charac_type == 'ex_speed':
            self.filter_rotation_ex()
            self.filter_torque_ex()                          
        
        
        #self.res_filtered.drop(self.res_filtered[self.res_filtered.index > self.index_last].index, 
        #                       inplace=True)       # drop tail of the the data
        self.res_filtered.dropna(inplace=True)     # ????????
                    
        # filtering fuel consumption
        self.filter_fuel()
        
        # def filter_first_last --- cut begining and end of each torque level
        self.filter_first_last()
               
        # visualise the result of the function job
        print('data were filtured --- def filter_data(self):')
        print(f'{len(self.res_filtered)} - amount or rowes in DataFrame after filtration')
        print('Filters used:')
        print(*[f'{key} - {val}' for key, val in self.dict_all_filtures.items()], sep='\n', end='\n\n')
        
    
    # used ONLY for load characteristics --- self.charac_type = 'load'
    def filter_rotation(self): 
        self.min_div1 = self.rotation - self.div_r   # def __init__(self, address, !!! rotation = 0 !!!) Ex.: 1600 Obc
        self.max_div1 = self.rotation + self.div_r
        self.res_filtered.drop(self.res_filtered[self.res_filtered['Obroty[obr/min]'] <= self.min_div1].index, 
                               inplace=True)
        self.res_filtered.drop(self.res_filtered[self.res_filtered['Obroty[obr/min]'] >= self.max_div1].index, 
                               inplace=True)
        
        self.dict_all_filtures['def filter_rotation(self):'] = True
        self.dict_all_filtures['    1.amount of rowes --- '] = len(self.res_filtered)
        
        
    # used ONLY for load characteristics --- self.charac_type = 'load'        
    def filter_torque(self, lst):
        self.lst_t = lst
        for tq in self.lst_t:
            self.div_min = tq - self.div_t
            self.div_max = tq + self.div_t
            self.index = self.res_filtered[(self.res_filtered['Moment obrotowy[Nm]'] >= self.div_min) & 
                             (self.res_filtered['Moment obrotowy[Nm]'] <= self.div_max)].index
            if len(self.index) > self.min_time_experiment:
                self.res_filtered.loc[self.index, 'Torque'] = tq        
                self.index_last = self.index[-1]   # used to drop the tail of the data in def filter_data(self):
                
        self.indices = []
        self.gr1 = self.res_filtered.groupby('Torque')  # group by 'Torque'
        self.dic1 = dict(list(self.gr1))
        for df in self.dic1.values():
            self.q_low = df.loc[:, 'Moment obrotowy[Nm]'].quantile(0.01)
            self.q_hi  = df.loc[:, 'Moment obrotowy[Nm]'].quantile(0.99)
            self.index = df[(df['Moment obrotowy[Nm]'] >= self.q_low) & 
                             (df['Moment obrotowy[Nm]'] <= self.q_hi)].index
            self.indices += list(self.index)
        self.drop_ind = list(set(self.res_filtered.index) - set(self.indices))
        self.res_filtered.drop(self.drop_ind, inplace=True)
        
        self.dict_all_filtures['def filter_torque(self, lst):'] = True
        self.dict_all_filtures['    2.amount of rowes --- '] = len(self.res_filtered)
        
        
    # used ONLY for external speed characteristics --- self.charac_type = 'ex_speed'
    def filter_rotation_ex(self):
        self.min_rot = self.rotation - self.div_r
        self.max_rot = self.rotation + self.div_r
        self.start_torque = 290
        
        #self.res_filtered.drop(self.res_filtered[self.res_filtered['Obroty[obr/min]'] >= self.max_rot].index, 
        #                       inplace=True)
        #self.res_filtered.drop(self.res_filtered[(self.res_filtered['Obroty[obr/min]'] <= self.min_rot) &
        #                                         (self.res_filtered['Moment obrotowy[Nm]'] <= self.start_torque)].index, 
        #                       inplace=True)
        
        self.res_filtered.drop(self.res_filtered[self.res_filtered['Moment obrotowy[Nm]'] <= self.start_torque].index, 
                               inplace=True)
               
        for rot in self.lst_rotation:
            self.div_min = rot - self.div_r
            self.div_max = rot + self.div_r
            self.index = self.res_filtered[(self.res_filtered['Obroty[obr/min]'] >= self.div_min) & 
                             (self.res_filtered['Obroty[obr/min]'] <= self.div_max)].index
            if len(self.index) > self.min_time_experiment:
                self.res_filtered.loc[self.index, 'Rotation'] = rot        
                self.index_last = self.index[-1]   # used to drop the tail of the data in def filter_data(self):
        
        self.dict_all_filtures['def filter_rotation_ex(self):'] = True
        self.dict_all_filtures['    3.amount of rowes --- '] = len(self.res_filtered)
        
           
    # used ONLY for external speed characteristics --- self.charac_type = 'ex_speed'    
    def filter_torque_ex(self):
                        
        self.indices = []
        self.get_dict()
        for df in self.dict_gr_res.values():
            self.q_low = df.loc[:, 'Moment obrotowy[Nm]'].quantile(0.01)
            self.q_hi  = df.loc[:, 'Moment obrotowy[Nm]'].quantile(0.99)
            self.index = df[(df['Moment obrotowy[Nm]'] >= self.q_low) & 
                             (df['Moment obrotowy[Nm]'] <= self.q_hi)].index
            self.indices += list(self.index)
        self.drop_ind = list(set(self.res_filtered.index) - set(self.indices))
        self.res_filtered.drop(self.drop_ind, inplace=True)
        
        self.dict_all_filtures['def filter_torque_ex(self):'] = True
        self.dict_all_filtures['    4.amount of rowes --- '] = len(self.res_filtered)
                     
                
    def filter_fuel(self):
        #self.fuel_means
        #np.random.seed(1)
        self.get_dict()
        self.count = 0
        for df, fuel_mean in zip(self.dict_gr_res.values(), self.fuel_means):
            self.lst_f = []
            self.min_max = [fuel_mean - self.div_f, fuel_mean + self.div_f]
            for val in df.loc[:, 'Zużycie paliwa średnie[g/s]']:
                if self.min_max[0] <= val <= self.min_max[1]:
                    self.lst_f.append(val)
                    self.count += 1
                else:
                    #self.lst_f.append(np.random.uniform(self.min_max[0], self.min_max[1])) 
                    self.lst_f.append(fuel_mean)
            self.res_filtered.loc[list(df.index), 'Zużycie paliwa średnie[g/s]'] = self.lst_f 
        
                
        self.dict_all_filtures['def filter_fuel(self):'] = True
        self.dict_all_filtures['    5.amount of rowes --- '] = len(self.res_filtered)
        print('self.count ---', self.count)
    
    # drop firsts and lasts in each group           
    def filter_first_last(self):        
        if self.charac_type == 'load':
            self.gr_fl = self.res_filtered.groupby('Torque')  # group by 'Torque'            
        elif self.charac_type == 'ex_speed':
            self.gr_fl = self.res_filtered.groupby('Rotation')  # group by 'Rotation'
        self.dic_fl = dict(list(self.gr_fl))
        for val in self.dic_fl.values():
            self.indices = list(val.index[:self.first_drop]) + list(val.index[-self.last_drop:])
            self.res_filtered.drop(self.indices, inplace=True)
        
        self.dict_all_filtures['def filter_first_last(self):'] = True
        self.dict_all_filtures['    6.amount of rowes --- '] = len(self.res_filtered)
    
    
    ''' !!!!!!!!!! TRANSFORMATIONS !!!!!!!!!!! '''        
    def transform(self):
        # dictionary of all transformations
        self.dict_all_transformations = {'def atmospheric_power_correction(self, Z=120000, displacement=4.5)': False,
                                         'def self.trans_temp_exhaust_gas(self):': False,
                                         'tr3': False,
                                         'tr4': False}
        self.res_transformed = self.res_filtered
        
        self.atmospheric_power_correction()
        
        # Temp. spalin 1...4/6[°C] --- get average
        self.trans_temp_exhaust_gas()
        
        # visualise the result of the function job
        print(f'data were transformed --- def transform(self):')
        print(f'{len(self.res_transformed)} - amount or rowes in DataFrame after transformation')
        print('Transformations applied:')
        print(*[f'{key} - {val}' for key, val in self.dict_all_transformations.items()], sep='\n', end='\n\n')
    
        
    def atmospheric_power_correction(self, Z=120000, displacement=4.5):
        self.df_cor = pd.DataFrame(columns=['q', 'r', 'qc', 'fm', 'fa', 'ac', 'Pref', 'Mref', 'Pref - P_orig [%]'], 
                                   index=self.res_transformed.index)
        self.df_cor.loc[:, 'q'] = Z * self.res_transformed.loc[:, 'Zużycie paliwa średnie[g/s]'] / \
                                    (4.5 * self.res_transformed.loc[:, 'Obroty[obr/min]'])
        self.df_cor.loc[:, 'r'] = self.res_transformed.loc[:, 'Ciś. pow. za turb.[Pa]'] / 10 / \
                                    self.res_transformed.loc[:, 'Ciśnienie atmosferyczne[hPa]']
        
        self.df_cor.loc[:, 'qc'] = self.df_cor.loc[:, 'q'] / self.df_cor.loc[:, 'r']
                
        self.df_cor.loc[:, 'fm'] = 0.036 * self.df_cor.loc[:, 'qc'] - 1.14
        self.lst_rep1 = list(self.df_cor[self.df_cor['qc'] <= 37.2]['fm'])
        self.lst_rep2 = list(self.df_cor[self.df_cor['qc'] >= 65]['fm'])
        self.df_cor['fm'].replace(self.lst_rep1, 
                                  [0.2]*len(self.lst_rep1), inplace=True)
        self.df_cor['fm'].replace(self.lst_rep2, 
                                  [1.2]*len(self.lst_rep2), inplace=True)
        
        self.df_cor.loc[:, 'fa'] = pow((99 / (self.res_transformed.loc[:, 'Ciśnienie atmosferyczne[hPa]'] / 10)), 0.7) \
                                    * pow(((self.res_transformed.loc[:, 'Temp. otoczenia[°C]'] + 273) / 298), 1.2)
        
        self.df_cor.loc[:, 'ac'] = pow(self.df_cor.loc[:, 'fa'], self.df_cor.loc[:, 'fm'])
        self.df_cor.loc[:, 'Pref'] = self.df_cor.loc[:, 'ac'] * self.res_transformed.loc[:, 'Moc[kW]']
        self.df_cor.loc[:, 'Mref'] = self.df_cor.loc[:, 'ac'] * self.res_transformed.loc[:, 'Moment obrotowy[Nm]']     
        
        self.df_cor.loc[:, 'Pref - P_orig [%]'] = (self.df_cor.loc[:, 'Pref'] \
                                                   - self.res_transformed.loc[:, 'Moc[kW]']) \
                                                    / self.res_transformed.loc[:, 'Moc[kW]'] \
                                                    * 100
        self.res_transformed.loc[:, 'Moc[kW]'] = self.df_cor.loc[:, 'Pref']
        self.res_transformed.loc[:, 'Moment obrotowy[Nm]'] = self.df_cor.loc[:, 'Mref']
        
        self.res_transformed.drop(['Ciśnienie atmosferyczne[hPa]', # del columns with atmospheric parameters.
                                   'Temp. otoczenia[°C]', 
                                   'Wilgotność względna[%]'],
                                  axis='columns', inplace=True)
                
        self.dict_all_transformations['def atmospheric_power_correction(self, Z=120000, displacement=4.5)'] = True
                
        
    def trans_temp_exhaust_gas(self):
        self.lst_gas = []  # lst_temp_exhaust_gas
        self.lst = ['Temp. spalin 1/6[°C]', 
                    'Temp. spalin 2/6[°C]', 
                    'Temp. spalin 3/6[°C]', 
                    'Temp. spalin 4/6[°C]']
        self.t_name = 'Temp. spalin 1...4/6 - srednia[°C]'
        self.res_transformed.loc[:, self.t_name] = 0
        for x in self.lst:
            if x in self.df.columns:
                self.lst_gas.append(x)
                self.res_transformed.loc[:, self.t_name] += self.res_transformed.loc[:, x]
        self.res_transformed.loc[:, self.t_name] = self.res_transformed.loc[:, self.t_name] / len(self.lst_gas)
        self.res_transformed.drop(self.lst_gas, axis='columns', inplace=True)
        
        self.dict_all_transformations['def self.trans_temp_exhaust_gas(self):'] = True
               
    
    def write_to(self, file_name='new_file'):
        self.res_transformed.to_csv(file_name) 
           
    
    ''' !!!!!!!! aditional Functions !!!!!!!!'''
    # group by 'Torque' or 'Rotation' --- and --- get dict
    def get_dict(self):
        if self.charac_type == 'load':
            self.grouped_res = self.res_filtered.groupby('Torque', sort=False)  # group by 'Torque'
            self.dict_gr_res = dict(list(self.grouped_res))         # get dict --- dictionary 1:df1, 25:df2,...
        elif self.charac_type == 'ex_speed':
            self.grouped_res = self.res_filtered.groupby('Rotation', sort=False)  # group by 'Rotation'
            self.dict_gr_res = dict(list(self.grouped_res))         # get dict --- dictionary 2400:df1, 2200:df2,...
            
    def add_ecology(self):
        self.get_dict()
        for key, val in self.dict_gr_res.items():
            self.columns_eco = ['CO', 'HC', 'LAMBDA', 'CO2', 'O2', 'NO', 'PM']
            self.key_index = list(self.df_eco[self.df_eco['OBR'] == key].index)
            self.res_transformed.loc[val.index[0], self.columns_eco] = self.df_eco.loc[self.key_index, 
                                                                                       self.columns_eco].values[0] 
        self.res_transformed.fillna(method='ffill', inplace=True)
    
    
    dic_fuel_prop = {'names': ['Сalorific value MJ/kg', 'Cetane number',
                               'Density at 15 °C', 'Viscosity at 40 °C '],
                     'Diesel': [43.1, 54.7, 832, 3.25],
                     'RME': [38, 56, 882.9, 4.454]
                          }
    def add_fuel_prop(self):
        self.res_transformed[self.dic_fuel_prop['names']] = self.dic_fuel_prop[self.fuel_name]
        
 
```

```python
class Graphs:
    ''' class --- different types of graphs
        1.  def first(): --- simple view on data '''
    #%matplotlib qt
    def first(self, df):         # Ex. "df" = t1.res_filtered
        self.table = df
        plt.style.use('bmh')
        self.dict_figures = {}
        for i in range(len(self.table.columns)): 
            self.fig, ax = plt.subplots()
            ax.plot(self.table.index, self.table.iloc[:, i])
            ax.set_title(self.table.columns[i])
            self.dict_figures[self.table.columns[i]] = self.fig
            
            # fig.savefig(f'plots/{fname}.png')
            # plt.close(fig)
                        
            
    def choose_column(self, dict_grouped, 
                      col='Zużycie paliwa średnie[g/s]'):   # Ex.  self.dict_gr_res ---- and ---- "col" column name
        self.table = dict_grouped
        self.column = col      
        plt.style.use('bmh')
        for key in self.table.keys():
            self.fig, ax = plt.subplots()
            ax.plot(self.table[key].index, self.table[key].loc[:, self.column])
            ax.set_title(f'{self.column} --- ex_speed - {key}')
            
    def describe_data(self):
        plt.style.use('bmh')
        
            
```
